## Read .csv from Multiple URLs with Pandas

```python
# link structure
## Link: root + season + league
"https://www.football-data.co.uk/mmz4281/" + "2122" + "/" + "E0" + ".csv"
```

```python
root='https://www.football-data.co.uk/mmz4281/'
```

## Loop

```python
links = ['E0', 'E2', 'E3', 'E1', 'EC']
allLinks = []
for link in links:
    url = root + "2122/" + link + ".csv"
    print("Trying:", url)
    eachLink = pd.read_csv(url, encoding="unicode_escape")
    allLinks.append(eachLink)
```

## Get multiple season csv file at a time

```python
for season in range(15,21):
    print(str(season) + str(season+1))
```

```python
links = ['E0', 'E2', 'E3', 'E1', 'EC']
all_csv_data = []

for link in links:
    for season in range(15, 21):  # range of seasons
        url = root + str(season) + str(season+1) + "/" + link + ".csv"
        each_csv = pd.read_csv(url, encoding="unicode_escape")

        each_csv.insert(1, 'season', season)  # make season a new column
        all_csv_data.append(each_csv)
```
