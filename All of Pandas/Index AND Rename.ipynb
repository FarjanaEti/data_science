## set_index() and sort_index() methods

```python
import pandas as pd
import numpy as np
```

```python
csv_data = pd.read_csv('StudentsPerformance.csv')
```

```python
csv_data
```

```python
new_index=np.arange(0,1000)
```

```python
import random
```

```python
random.shuffle(new_index)
```

```python
new_index
```

```python
csv_data['new_index_col']=new_index
```

```python
csv_data
```

### set new index by replacing original one

```python
csv_data
```

```python
# here ('column you want to make index')
csv_data.set_index('new_index_col', inplace=True)
csv_data
```

```python
csv_data.sort_index()
```

```python
csv_data.sort_index(ascending=False)
```

# Rename Columns and Index with rename()

### Rename Column

```python
csv_data=csv_data.rename(columns={'gender':'Gender'})
```

```python
csv_data
```

```python
## rename multiple columns
csv_data.rename(columns={'lunch': 'lunch type', 'reading score':'reading mart'}, inplace=True)
```
