## Add New Column with assign() and insert()

```python
csv_data.head()
```

#### assign() is use to assign multiple column in one line code 
#### and overwrite the value of the existing columns

### Syntax--> df = df.assign(Name=values, Name=values2)
### np.random gives a array so we ahve to assign them a index using pd.Series

```python
NewC1=np.random.randint(1, 100, size=1000)
NewC2=np.random.randint(1,100,size=1000)
```

```python
series1=pd.Series(NewC1, index=np.arange(0,1000))
series2=pd.Series(NewC2, index=np.arange(0,1000))
```

```python
csv_data=csv_data.assign(NewC1=series1, NewC2=series2)
```

```python
csv_data
```
