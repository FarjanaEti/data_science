##  value_counts() method

```python
import pandas as pd
```

```python
csv_data=pd.read_csv('StudentsPerformance.csv')
```

```python
csv_data.head()
```

```python
len(csv_data['gender'])
```

```python
csv_data['gender'].count()
```

### count by category

```python
csv_data['gender'].value_counts()
```

```python
csv_data['parental level of education'].value_counts()
```

```python
## finding the parcentage (relative frequency)
```

```python
csv_data['parental level of education'].value_counts(normalize=True)
```

```python
csv_data['gender'].value_counts(normalize= True)
```
