```python
import pandas as pd
```

```python
fifa_data=pd.read_csv('players_20.csv')
```

```python
fifa_data
```

```python
## set a certain column as main index
##fifa_data.set_index('age', inplace=True)
fifa_data.set_index('short_name', inplace=True)
```

```python
fifa_data.head(5)
```

```python
### data=data[[]] willl take only given column from original dataframe and make a new data frame
```

```python
fifa_data=fifa_data[['long_name','age',  'dob', 'height_cm', 'weight_kg','nationality', 'club']]
```

```python
fifa_data
```

## Selecting with a single value
### loc[row_label, column_label]

```python
## all data of a row(messi)
fifa_data.loc['L. Messi']
```

```python
fifa_data.loc['L. Messi','height_cm']
```

```python
fifa_data.loc['Cristiano Ronaldo', 'weight_kg']
```

```python
# get all rows inside the 'height_cm' column
fifa_data.loc[:, 'height_cm']
```

```python
# get all columns that correspond to the index 'L. Messi'
fifa_data.loc['Neymar Jr', :]
```

### Selecting with a list of values

```python
## getting all data of more than pone player
fifa_data.loc[['L. Messi', 'Cristiano Ronaldo']]
```

```python
## getting their height
fifa_data.loc[['L. Messi', 'Cristiano Ronaldo'],'height_cm']
```

```python
## getting multiple column data of one player
fifa_data.loc['Neymar Jr',['height_cm',	'weight_kg']]
```

```python
## getting multiple data of multiple player
fifa_data.loc[['L. Messi', 'Cristiano Ronaldo'], ['height_cm',	'weight_kg','age']]
```

### Selecting a range of data with a slice

```python
## slice with column
players = ['L. Messi', 'Cristiano Ronaldo']
fifa_data.loc[players, 'age':'nationality']
```

```python
## slice with row
columns = ['age', 'dob', 'height_cm', 'weight_kg']

# get top1 and top10 player name
 #fifa_data.index[:10]
fifa_data.loc['L. Messi':'E. Hazard', columns]
```

## Selecting with conditions

```python
## selecting player with column but one column with a condition
columns = ['age', 'dob', 'height_cm', 'weight_kg']
fifa_data.loc[fifa_data['weight_kg']> 80, columns]
```

```python
fifa_data.loc[(fifa_data['height_cm']>180) & (fifa_data['nationality']== 'Argentina'),:]
```
