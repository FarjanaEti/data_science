```python
import pandas as pd
```

```python
df=pd.read_csv('players_20.csv')
df.head(5)
```

## Selecting with a single value

```python
df.set_index('short_name', inplace=True)
df.head(5)
```

```python
df=df[['long_name','age',  'dob', 'height_cm', 'weight_kg','nationality', 'club']]
df
```

```python
## get the set-indexed 0 index 3 columns value
df.iloc[0,3]
```

```python
df.iloc[1,4]
```

```python
##  giving nothing means all
## row : column
df.iloc[0: 2]
```

```python
## no comma rows only and all column as nothing is required
##Select the first 3 rows, and include ALL columns.
df.iloc[: 3]
```

```python
## Select all rows but only the 3rd column.
df.iloc[:,2]
```

```python
## 3rd row all column
df.iloc[2, : ]
```

## Selecting with a list of values

```python
## before comma all rows even : means all or 
## even withh [[rows],[column]]  [[0,1]]--> here it means 0 and 1 rows this comma is for row 
```

```python
df.iloc[[0,1]]
```

```python
df.iloc[[1,2], 3]
```

```python
df.iloc[0, [3,4]]
```

```python
df.iloc[[1,2],[3,4]]
```

## Selecting a range of data with a slice

```python
## 1:6 col start : col end(n-1)
rows=[0,2]
df.iloc[rows, 1:6]
```

```python
cols=[1,2,3,4,5]
df.iloc[0:5, cols]
```

## Slice and condition

```python
cols=[1,2,3]
df.iloc[list(df['height_cm']>180),cols]
```

```python
cols=[1,3,4,5]
df.iloc[list((df['height_cm']>180) & (df['age']>30)),cols]
```
