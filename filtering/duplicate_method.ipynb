### Task: Find the cheapest and most expensive laptop of each company using the sort_values() and duplicated() methods (keep "first", "last" and False)

```python
## sort data company and price
sort_data=laptop_data.sort_values(['Company', 'Price_euros'])
sort_data
```

```python
laptop_data.value_counts('Company')
```

```python
duplicated_company=laptop_data.duplicated('Company')
##duplicated_company this will show true and false
laptop_data[duplicated_company]
```

```python
## ~ shows all uniqe data no duplicated
laptop_data[~duplicated_company]
```

```python
##
laptop_data[~ duplicated_company][['Company','Price_euros']]
```

```python
laptop_data[~duplicated_data].value_counts('Company')
```

## keep= first

```python
duplicated_last = laptop_data.duplicated('Company', keep='last')
```

```python
duplicated_false= laptop_data.duplicated('Company', keep=False)
```
