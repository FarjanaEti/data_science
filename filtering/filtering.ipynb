```python
import pandas as pd
```

```python
laptop_data = pd.read_csv("laptop_price.csv", encoding="latin1")
```

```python
## Finding which row has "Apple" laptop
laptop_data['Company'] == "Apple"
```

```python
## Getting all the dataFrame that have "Apple" data
laptop_data[laptop_data["Company"]=="Apple"]
```

```python
##finding the number of company having apple or HP
laptop_data[laptop_data["Company"]=="HP"].value_counts("Company")
```

```python
laptop_data[laptop_data["Company"]=="Dell"].value_counts("Company")
```

##### Not Equal

```python
## Finsing all the company exept HP
laptop_data["Company"]!="HP"
```

```python
## number ohh other company without Dell
laptop_data[laptop_data["Company"]!= "Dell"].value_counts("Company")
```

```python
laptop_data[laptop_data["Price_euros"]>2000]
```

## Multiple condition

```python
laptop_data[(laptop_data["Company"]=="Apple") & (laptop_data["Price_euros"]>2000) ]
```

```python
laptop_data[(laptop_data["Company"]== "Dell") | (laptop_data["Company"]=="HP")]
```

## Multiple condition using .isin()

```python
laptop_data[laptop_data["Company"].isin(["Apple",'HP','Dell'])]
```

### sort after filter

```python
laptop_data[laptop_data["Company"]=="HP"].sort_values(by="Price_euros", ascending=False)
```
