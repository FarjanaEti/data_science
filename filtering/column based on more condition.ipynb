## More Than 2 Choices np.select().ipynb

```python
## create condition and values
# create "conditions" and "values" variables
conditions = [
    (laptop_data['Price_euros']>3000),
    (laptop_data['Price_euros']>2000) & (laptop_data['Price_euros']<=3000),
    (laptop_data['Price_euros']>800) & (laptop_data['Price_euros']<=2000),
    (laptop_data['Price_euros']<=800),
]
choices = ['Budget', 'Mid-range', 'Premium', 'Luxury']
```

```python
## default= unknown if none of the conditions in your list match.
laptop_data['Price_tier'] = np.select(conditions, choices,  default='Unknown')
```

```python
laptop_data
```

```python
laptop_data.value_counts('Price_tier')
```
