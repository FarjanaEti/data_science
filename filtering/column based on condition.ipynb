```python
import pandas as pd
```

```python
laptop_data = pd.read_csv("laptop_price.csv", encoding="latin1")
```

```python
laptop_data.head(3)
```

```python
import numpy as np
```

```python
laptop_data["Price_euros"]>2000
```

```python
## creating array from dataframe
np.where(laptop_data["Price_euros"]>2000)
```

```python
## it can take two argument if true and false
np.where(laptop_data["Price_euros"]> 2000, "Dami", "kom dami")
```

```python
### Creating new column from the price data
laptop_data["New-col"]=np.where(laptop_data["Price_euros"]> 2000, "Dami", "kom dami")
```

```python
laptop_data.head(10)
```

```python
laptop_data.value_counts("New-col")
```
