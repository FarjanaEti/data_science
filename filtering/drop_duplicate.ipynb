```python
import pandas as pd
```

```python
laptop_data=pd.read_csv("laptop_price.csv", encoding="latin1")
```

```python
## This will give all the uniqe data without dupliocated company data
laptop_data.drop_duplicates(['Company'])
```

```python
## For ensure that company is not duplicated choose company and price column
laptop_data.drop_duplicates(['Company'])[['Company','Inches']]
```

```python
## sort dataFrame in ascending
## cheapest first and expecsive last
sort_data=laptop_data.sort_values(['Company', 'Price_euros'])
sort_data
```

```python
## keep=first it get it by default
## this will keep the first occurrence of each Company and drop the rest.
laptop_data.drop_duplicates(['Company']) [['Company', 'Price_euros']]
```

```python
## laptop_data.drop_duplicates(['Company']) this will give all dataframe
## so basically this two is same
## to get cheapest first sort it
sort_data.drop_duplicates(['Company'], keep='first')[['Company', 'Price_euros']]
```

```python
sort_data.drop_duplicates(['Company'], keep='last') [['Company', 'Inches']]
```
